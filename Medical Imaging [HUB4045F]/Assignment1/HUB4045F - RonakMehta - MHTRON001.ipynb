### Solution to Question 1: 

Consider the function ${f}$ , whose domain is the plane, the Radon transform of ${f}$  is defined, for each pair of real numbers ${(t, \theta)}$, by
\begin{align}
\mathcal{R} f(t,\theta ) & := \int_{l_{t,\theta}}^{} f ds \\
& = \int_{s=-\infty}^{\infty} f(t \cos(\theta) - s\sin(\theta),t \sin(\theta) + s\cos(\theta)) ds.
\end{align} 

Note that both ${f}$ and ${\mathcal{R} f}$ are functions. 

${f}$ is a function of the Cartesian coordinates ${x}$  and  ${y}$ while ${\mathcal{R} f}$ is a function of the polar coordinates ${t}$  and ${\theta}$. 

For each choice of ${f}$, ${t}$, and ${\theta}$, ${\mathcal{R} f(t,\theta)}$ is a number. 

In the integral on the right, the variable of integration is ${s}$ , while the values of ${t}$  and ${\theta}$ are preselected and  so should be treated as constants when evaluating the integral.

We can visualize the Radon transform of a given function by treating ${t}$  and ${\theta}$ as _rectangular coordinates._ We  can also restrict our natural domain of ${\mathcal{R} f(t,\theta)}$ such that it is the set of absolutely integrable functions.

${\int_{s=-\infty}^{\infty}\mid (f(t \cos(\theta) - s\sin(\theta),t \sin(\theta) + s\cos(\theta))\mid ds}$

This is exactly what is happening in the gif given where the image to the left is a sample through which X-rays are being passed at various angles (from 0 to 180 degrees) and the values of Radon transform being depicted according to their brightness on a continuum of grey values on the right (where usually the value 0 represents the color black, 0:5 represents a neutral grey, and the value 1 represents a white). I.e. a __sinogram__ which depicts all of the data generated by the X-ray emission/detection machine for this given gif slice of the brain sample.

#### _Consider a ${3}$ x ${3}$ grid in which each of the nine squares is shaded either black or white_. 
Find all possible grids for which  which Rows 1 and 2 each have two white squares and Row 3 has one white square; Columns 1 and 3 each have one white square and Column 2 has three white squares.

```python
# SOLUTION TO QUESTION 2

import numpy as np
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3, squeeze=False, figsize=(12, 2))

# Counter for number of grids which fit constraints
counterA = 0;

# Assign values to grid and test
for gridNum in range (0, 512):
    
# Create grid
    row1Whites = row2Whites = row3Whites = col1Whites = col2Whites = col3Whites = 0;
    curGrid = [[0, 0, 0], [0, 0, 0], [0, 0, 0]];
    temp = gridNum;
    if ((temp - 256) >= 0 ):
        curGrid[0][0] = 1;
        temp -= 256;
    if ((temp - 128) >= 0 ):
        curGrid[0][1] = 1;
        temp -= 128;
    if ((temp - 64) >= 0 ):
        curGrid[0][2] = 1;
        temp -= 64;
    if ((temp - 32) >= 0 ):
        curGrid[1][0] = 1;
        temp -= 32;
    if ((temp - 16) >= 0 ):
        curGrid[1][1] = 1;
        temp -= 16;
    if ((temp - 8) >= 0 ):
        curGrid[1][2] = 1;
        temp -= 8;
    if ((temp - 4) >= 0 ):
        curGrid[2][0] = 1; 
        temp -= 4;
    if ((temp - 2) >= 0 ):
        curGrid[2][1] = 1;
        temp -= 2;
    if ((temp - 1) >= 0 ):
        curGrid[2][2] = 1
        
# Test Constraints 
    for col in range (0, 3):
        if (curGrid[0][col] == 1):
            row1Whites += 1;
        if (curGrid[1][col] == 1):
            row2Whites += 1;
        if (curGrid[2][col] == 1):
            row3Whites += 1;
    for row in range (0, 3):
        if (curGrid[row][0] == 1):
            col1Whites += 1;
        if (curGrid[row][1] == 1):
            col2Whites += 1;
        if (curGrid[row][2] == 1):
            col3Whites += 1;
    if (row1Whites == 2 and row2Whites == 2 and row3Whites == 1 and col1Whites == 1 and col2Whites == 3 and col3Whites == 1):
        
# Count and display each grid which satisfies constraints
        counterA += 1;
        print(curGrid);
        image = np.array(curGrid);
        axs[0,(counterA-1)].imshow(image, cmap=plt.cm.Greys_r)
        axs[0,(counterA-1)].set_yticklabels([])
        axs[0,(counterA-1)].set_xticklabels([])
        
print ("Possible configurations, as shown above are %(num)d" % {"num": counterA});
print("Note that 0 means a black square while 1 means a white square");
```

### Solution to Question 3:

A line ${\boldsymbol l}$ in the plane  is a set of points that satisfies an equation of the form
 <center> ${ax + by =c}$, where ${a^2 + b^2 \neq 0}$ </center>
 
However knowing ${a^2 + b^2 \neq 0}$ we can rewrite the standard equation of a line ${\boldsymbol l}$ as 
${\frac{a}{\sqrt{a^2+b^2}}x + \frac{b}{\sqrt{a^2+b^2}}y = \frac{c}{\sqrt{a^2+b^2}}}$ 

Letting ${t=(\frac{c}{\sqrt{a^2+b^2}})}$, we can rewrite ${\boldsymbol l}$ as: $
{xcos\theta + ysin\theta =t}$

Comparing the  equations give us that $
{cos\theta ={\frac{a}{\sqrt{a^2+b^2}}}}$
and ${sin\theta = {\frac{b}{\sqrt{a^2+b^2}}}}$

And since ${tan\theta = {\frac{sin\theta}{cos\theta}}}$, we get ${tan\theta = {\frac{b}{a}}}$

Then given the line with equation ${\sqrt{3}x+y=4}$, 

We can see that a=${\sqrt{3}}$, b=1 and c=4

Substituting the values in the above equations for tan and t yields ${tan\theta ={\frac{1}{\sqrt{3}}}}$
and
${t=\frac{4}{\sqrt{{\sqrt{3}}^2+1^2}} = \frac{4}{\sqrt{3+1}} = \frac{4}{2} = 2}$

${\theta}$ is found to be 30&deg; or ${\frac{\pi}{6}}$

Hence, we can conclude by saying ${l_{2,\frac{\pi}{6}}}$ is the same as the line with equation ${\sqrt{3}x+y=4}$

### Question 5: [10 marks]
Provide a proof of the central slice theorem, starting with the expression ${\mathcal{F}(\mathcal{R}f)(S,\theta)}$  and ending with ${\mathcal{F}_2 f (S \cos (\theta), S \sin(\theta))}$



### Solution to Question 5

The interaction between the Fourier transform and the Radon transform is expressed in an equation known as the __central slice theorem__  

${\mathcal{F}(\mathcal{R}f)(S,\theta)}$ ${= \int_{-\infty}^{\infty}(\int_{-\infty}^{\infty} f(t \cos\theta - s\sin\theta,t \sin\theta + s\cos\theta)ds) e^{-iSt}dt}$

Re-arranging the integral term gives us:

${= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} f(t \cos\theta - s\sin\theta,t \sin\theta + s\cos\theta) e^{-iSt}dsdt}$

But from parameterization of a line, we know that:

${x = t \cos(\theta) - s \sin(\theta)}$ , 

${y = t \sin(\theta) + s \cos(\theta)}$ 

${t = x \cos(\theta) + y \sin(\theta)}$

Therefore, substituting these values in the above integral gives us:
<center>${\mathcal{F}(\mathcal{R}f)(S,\theta)}$ ${= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} f(x, y) e^{-iS(x\cos\theta + y\sin\theta)}dxdy}$</center>

Notice that this is the 2D Fourier transform of ${f(x, y)}$. Fourier Slice Theorem relates 1D Fourier Transform of the projection with 2D Fourier Transform of the original image

Changing from cartesian to polar coordinates require a change of variable where ${x=S\cos\theta}$ and ${y=S\sin\theta}$

Thus, dxdy = $\mid{S}\mid$ dS d$\theta$


<center>${\mathcal{F}(\mathcal{R}f)(S,\theta)}$ ${= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(S\cos\theta,S\sin\theta) e^{iS(x\cos\theta + y\sin\theta)}\mid{S} \mid dS d\theta}$</center>

Hence, for any suitable function ${f}$ defined in the plane and all real numbers ${S}$ and ${\theta}$: 

<center><font color='blue'>${{\mathcal{F}(\mathcal{R}f)(S,\theta)} = \mathcal{F}_2 f (S \cos (\theta), S \sin(\theta))}$</font> </center>

NB: The symbols ${\mathcal{F}}$ and  ${\mathcal{F}_2}$ are used to denote the  ${1}$D and ${2}$D Fourier transforms, respectively. The Radon transform is denoted by${\mathcal{R}_2}$. The function ${f}$, representing, say, an X-ray attenuation coefficient, is a function of ${2}$D Cartesian coordinates.

We then have the Jacobian determinant: 

\begin{equation}
\begin{vmatrix}
\frac{\partial X}{\partial S} &  \frac{\partial X}{\partial \theta}\\

\end{vmatrix} =  \mid {S} \mid
\end{equation} 

Thus, dXdY = $\mid{S}\mid$ dS d$\theta$

Hence, 

$f(x,y)$ ${= \frac{1}{4\pi^2}\int_{0}^{\pi}\int_{-\infty}^{\infty} F_{2} f(S\cos\theta,S\sin\theta) e^{iS(x\cos\theta + y\sin\theta)}\mid{S} \mid dS d\theta}$

From Central Slice theorem; $F_{2}$$f(S\cos\theta,S\sin\theta)$ = $F(Rf)(s,\theta)$

$f(x,y)$ ${= \frac{1}{4\pi^2}\int_{0}^{\pi}\int_{-\infty}^{\infty} F(Rf)(s,\theta) e^{iS(x\cos\theta + y\sin\theta)}\mid{S} \mid dS d\theta}$
